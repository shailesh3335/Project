##Task 2 - Descriptive Statistics üìä

- **Calculate measures of central tendency (mean, median, mode) and measures of dispersion (variance, standard deviation) for key variables. Identify and handle outliers if necessary.**


**Deliverables**:

- **Descriptive statistics that reveal the central tendencies, variations, and potential outliers in the dataset.**:

  

```python
# Calculate measures of central tendency and dispersion for "Income"

# Central Tendency and Dispersion for "Income"
mean_income = df['Income'].mean()
median_income = df['Income'].median()
mode_income = df['Income'].mode().iloc[0]  # Mode might return multiple values, so we select the first one
variance_income = df['Income'].var()
std_deviation_income = df['Income'].std()

# Identify and handle outliers (optional)
# You can use various methods such as z-scores or IQR to detect and handle outliers.

# For example, to detect outliers for "Income" using z-scores:
z_scores_income = (df['Income'] - mean_income) / std_deviation_income
outliers_income = df[abs(z_scores_income) > 3]  # Define a threshold (e.g., 3 standard deviations)

# Handle outliers by removing them from the DataFrame
df = df[abs(z_scores_income) <= 3]
df
```

```python
# Visualize the distribution of "Income" (optional)
plt.hist(df['Income'], bins=20, color='lightgreen')
plt.xlabel('Income')
plt.ylabel('Count')
plt.title('Income Distribution')
plt.show()

```

```python
# Print the calculated statistics and the DataFrame
print("Measures of Central Tendency for Income:")
print(f"Mean Income: {mean_income}")
print(f"Median Income: {median_income}")
print(f"Mode Income: {mode_income}")

print("\nMeasures of Dispersion for Income:")
print(f"Variance Income: {variance_income}")
print(f"Standard Deviation Income: {std_deviation_income}")

print("\nOutliers for Income (if any):")
print(outliers_income)

print("\nDataFrame after handling income outliers:")
print(df)
```

## Task 3 - Probability Distributions üé≤

- **Identify variables that could follow specific probability distributions (e.g., Binomial, Normal). Calculate probabilities and expected values based on these distributions.**



**Deliverables**:

- **Determination of suitable probability distributions for relevant variables and corresponding calculated probabilities and expected values.**:

  

```python
from scipy.stats import binom, norm

# Identify variables and distributions
# Example: NumDealsPurchases might follow a Binomial distribution
binom_params = (df['NumDealsPurchases'].nunique() - 1, df['NumDealsPurchases'].mean() / df['NumDealsPurchases'].nunique())

# Calculate probabilities and expected values
probabilities = binom.pmf(np.arange(df['NumDealsPurchases'].min(), df['NumDealsPurchases'].max() + 1), *binom_params)
expected_value = binom.mean(*binom_params)

# Display results
print("Probabilities:\n", probabilities)
print("\nExpected Value:\n", expected_value)
```
