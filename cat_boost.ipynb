```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("/content/train.csv")
test = pd.read_csv("/content/test.csv")
sample_sub = pd.read_csv("/content/sample_submission_V9Inaty.csv")
```

```python
train1 = pd.get_dummies(train.iloc[:,2:])
```

```python
test1 = pd.get_dummies(test.iloc[:,2:])
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy="mean")
```

```python
train_2 = imp.fit_transform(train1)
test_2 = imp.fit_transform(test1)
train_1 = pd.DataFrame(train_2,columns=train1.columns)
test_1 = pd.DataFrame(test_2,columns=test1.columns)
```

```python
x = train_1.drop(["Purchase"],axis=1)
y = train_1["Purchase"]
```

```python
from catboost import CatBoostRegressor
```

```python
parameters = {'depth'         : [9,10],
              'learning_rate' : [0.25,0.26,0.3],
              'iterations'    : [200,400,500,]
                 }
```

```python
from sklearn.model_selection import GridSearchCV,KFold
```

```python
model = CatBoostRegressor()
```

```python
kfold = KFold(n_splits=5,shuffle=True,random_state=2022)
```

```python
cv = GridSearchCV(model, param_grid=parameters,
                  scoring='r2',verbose=3)
cv.fit(x,y)
```

```python
print(cv.best_params_)

print(cv.best_score_)
```

```python
y_pred =  cv.predict(test_1)
```

```python
y_pred
```

```python
sample_sub["Purchase"]= y_pred
```

```python
sample_sub.to_csv("/content/Sample catboost3.csv",index=False)
```
